```python
def even_odd(num):
    if num % 2 == 0:
        return "even number"
    else:
        return "odd number"
print(even_odd(5))
```

```python
def largest_of_three_numbers(x,y,z):
    print(f"x = {x}, y = {y}, z = {z}\n")
    if x>y and x>z:
        return f"x = {x} is large."
    elif y>x and y>z:
        return f"y = {y} is large."
    else:
        return f"z = {z} is large."

print(largest_of_three_numbers(1,5,9))
```

```python
def vowels_count(vowels,s):
    count = 0
    for i in s:
        if i in vowels:
            count+=1
    return count
        
vowels = "AEIOUaeiou"
s = "Alphabet"
print(f"Count of vowels = {vowels_count(vowels,s)}")
```

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)
print(factorial(5))
```

```python
def printNumbers(n):
    if n == 0:
        return

    printNumbers(n - 1)

    print(n, end=' ')

if __name__ == "__main__":
    n = 5
    print("Numbers from 1 to", n, ": ", end="")
    printNumbers(n)
    print()
```

```python
def prime(n):
    if n <= 1:
        return False
    for i in range(2,int(n ** 0.5)+1):
        if n % i == 0:
            return False
    return True
print(prime(7))
    
```

```python
def reverse_string(s):
    return s[::-1]
print(reverse_string("python"))
```

```python
def is_palindrome(s):
    return s == s[::-1]
print(is_palindrome("liril"))
```
