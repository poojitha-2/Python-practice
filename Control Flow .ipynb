### Conditional statements(if,elif,else)

```python
num = int(input("Enter a number :"))
if num > 0:
    print("positive number")
elif num < 0:
    print("Negative number")
elif num == 0:
    print("zero")
```

```python
num = int(input("Enter number :"))
if num % 2 == 0:
    print("even number")
else:
    print("odd number")
```

```python
num = int(input("enter number: "))
if num > 10:
    print("Greater than 10")
elif num < 10:
    print("Less than 10")
else:
    print("Equal to 10")
```

```python
age = int(input("Enter age :"))
if age >= 18:
    print("Eligible to vote.")
else:
    print("Not eligible to vote.")
```

```python
marks = 87
if marks >= 40:
    print("Pass")
else:
    print("Fail")
```

```python
num1 = int(input("Enter number :"))
num2 = int(input("Enter number :"))
if num1 > num2:
    print(f"{num1} is greater than {num2}")
else:
    print(f"{num2} is greater than {num1}")
```

```python
num1 = int(input("Enter number :"))
num2 = int(input("Enter number :"))
num3 = int(input("Enter number :"))
if num1 > num2 and num1 > num3:
    print(f"{num1} is greater")
elif num2 > num1 and num2 > num3:
    print(f"{num2} is greater")
else:
    print(f"{num3} is greater")
```

```python
num = int(input("enter number:"))
if num % 3 == 0 and num % 5 == 0:
    print("Divisible by 3 and 5")
elif num % 3 == 0:
    print("Divisible by 3")
elif num % 5 == 0:
    print("Divisible by 5")
else:
    print("Not divisible by 3 and 5")
```

```python
x = 10
if x > 5:
    print("Hello")
if x > 8:
    print("World")
```

```python
marks = int(input("Enter marks: "))
if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 60:
    print("Grade C")
elif marks >= 40:
    print("Grade D")
else:
    print("Fail")
```

```python
char = input("Enter a character: ")
vowels = "aeiouAEIOU"
if len(char) == 1 and char.isalpha():
    if char in vowels:
        print(f"{char} is a vowel")
    else:
        print(f"{char} is a consonant")
else:
    print("Enter single letter")
```

```python
num = int(input("Enter a number: "))
if num > 0:
    if num % 2 == 0:
        print("positive and even")
    else:
        print("positive and odd")
else:
    print("Negative")
```

### LOOPS

```python
#for loop

for i in range(1,11):
    print(i)
```

```python
for i in range(1,21):
    if i % 2 == 0:
        print(i,end = " ")
```

```python
str = "Python"
for i in str:
    print(i)
```

```python
for i in range(1,6):
    print(f"square of {i} = {i*i}")
```

```python
for i in range(10,0,-1):
    print(i)
```

```python
for i in range(1,11):
    print(f"5 * {i} = {5*i}\n")
```

```python
s=0
for i in range(1,6):
    s = s+i
print(s)
```

```python
vowels = "aeiouAEIOU"
str = "AutUmn"
count = 0
for i in str:
    if i in vowels:
        count+=1
print(count)
```

```python
i = 1
while i <= 5:
    print(i)
    i+=1
    
```

```python
i = 5
while i>=1:
    print(i)
    i-=1
```

```python
while True:
    num = int(input("Enter number: "))
    if num == 0:
        break
```

```python
for i in range(11):
    if i == 7:
        break
    print(i)
        
```

```python
for i in range(10):
    if i == 5:
        continue
    print(i)
```

```python
for i in range(1,5):
    print("*" * i)
```

```python
for i in range(1,5):
    for j in range(1,i+1):
        print(j,end = "")
    print()
```

```python
num = "12345"
for i in range(len(num),0,-1):
    print(i,end = "")
```
